```python
## load all libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from glob import glob
import librosa as lr
import librosa.display
```

```python
data_dir = 'C:/Users/spatra/Desktop/sounds/Actor_01'
file_data = {}
Main_dir = 'C:/Users/spatra/Desktop/sounds/Actor_'
```

```python
####Reading all files
#### Crate a dictionary with file name and length of the files

for files in range(1,25):
    if len(str(files))==1:
        data_dir = Main_dir
        data_dir = data_dir+str('0')+str(files)
    else:
        data_dir = Main_dir
        data_dir = data_dir+str(files)
#    print(data_dir)
        audio_files = glob(data_dir+'/*.wav')
        for i in range(len(audio_files)):
            x= audio_files[i][31:]
            audio,sfreq=lr.load(audio_files[i],sr=None)
            file_data[x]= len(audio)/sfreq
```

```python
### convert to data Frame
data_df=pd.DataFrame()
for keys,values in file_data.items():
    data_df.at[keys,'lengthoffile']=values
```
